```python
import pandas as pd 
from bs4 import BeautifulSoup
import re 
import os
from characterCleaner import simple_clean

def filterDate(soup):
    # this function is for Stage II filtering 
    dateStr = soup.find_all('date')[1].string
    intDates = re.findall(r'\d+',dateStr)
    if len(intDates) != 0:
        for d in intDates:
            if int(d) in range(1580,1640+1):
                return (True,int(d))
    return (False,None)

# Stage I functions
def authors(soup):
    authors = soup.find_all('author')
    auth = ''
    for a in authors:
        if auth != '':
            if not re.search(a.get_text(),auth):
                auth = auth + '; ' + a.get_text()
        else: 
            auth += a.get_text()
    if auth == '':
        auth = 'Anonymous'
    return auth

def pub(soup):
    publisher = soup.find_all('publisher')[1].string
    pubplace = soup.find_all('pubplace')[1].string
    keywords = soup.find_all('keywords')
    pubStr = publisher + '\n' + pubplace
    if len(keywords) != 0:
        keywords = soup.find_all('keywords')[0].get_text()
        pubStr = pubStr + '\n' + keywords 
    return pubStr

def date(soup):
    dateStr = soup.find_all('date')[1].string
    # keep dateStr unaltered if it is a date range, e.g., [between XXXX and XXXX] or [between XXXX-XXXX]
    estimates = re.search('between',dateStr)
    if estimates != None:
        return dateStr
    # otherwise, convert dateStr to a single date 
    intDates = re.findall(r'\d{4}',dateStr)
    if len(intDates) != 0:
        for d in intDates:
            if int(d) in range(1470,1800):
                return int(d)
    return 'Date Unknown'
```
